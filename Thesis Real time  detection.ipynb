```python
!pip install ultralytics
# Calculate and display mAP, Recall, Accuracy, Precision
metrics = ultralytics.metrics.compute_metrics(
    predictions='runs/detect/train/predict.json',
    ground_truth='vehicle_dataset/valid/labels',
    nc=5,
    names=['car', 'threewheel', 'bus', 'truck', 'motorbike', 'van']
)

print(f"mAP: {metrics['mAP']:.4f}")
print(f"Recall: {metrics['recall']:.4f}")
print(f"Precision: {metrics['precision']:.4f}")
print(f"Accuracy: {metrics['accuracy']:.4f}")

# Visualize the metrics using seaborn
sns.set(style="whitegrid")
metrics_df = pd.DataFrame(metrics, index=[0])
sns.barplot(data=metrics_df, palette="viridis")
plt.title("Model Performance Metrics")
plt.show()
```
